```python
import numpy as np
import matplotlib.pyplot as plt
```

importing required modules

*   numpy : calculations
*   matplotlib.pyplot : plotting graph



```python
x = np.arange(-10,10,0.001)
y = np.exp(x)
```

x values between -10 & 10 with step length of 0.001

y is exponential fnctn of x i.e y=e^x

```python
plt.grid()
plt.plot(x,y, "g")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('Exponential Curve')
plt.show()
```

Plotting y values against x in green coloured cont. line

Making a grid

Labelling the plot and giving command to display it.
